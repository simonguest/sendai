## üìä Quick Overview: India at a Glance

Let's start by looking at some basic facts about India's states and union territories!

```python
# Display basic information
print("üó∫Ô∏è INDIA'S ADMINISTRATIVE DIVISIONS")
print("=" * 40)

# Count by type
type_counts = india_df['Type'].value_counts()
print(f"States: {type_counts['State']}")
print(f"Union Territories: {type_counts['UT']}")
print(f"Total: {len(india_df)}")

print("\nüåç REGIONAL DISTRIBUTION")
print("=" * 25)
region_counts = india_df['Region'].value_counts()
for region, count in region_counts.items():
    print(f"{region}: {count}")

# Create a simple visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Pie chart for States vs UTs
colors1 = ['#FF6B6B', '#4ECDC4']
ax1.pie(type_counts.values, labels=type_counts.index, autopct='%1.1f%%', 
        colors=colors1, startangle=90)
ax1.set_title('States vs Union Territories', fontsize=14, fontweight='bold')

# Bar chart for regional distribution
colors2 = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7']
bars = ax2.bar(region_counts.index, region_counts.values, color=colors2)
ax2.set_title('Distribution by Region', fontsize=14, fontweight='bold')
ax2.set_xlabel('Region')
ax2.set_ylabel('Number of States/UTs')
ax2.tick_params(axis='x', rotation=45)

# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + 0.1,
             f'{int(height)}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

print("\nüí° Did you know?")
print("India has more states than most countries have provinces!")
print("Each state has its own government and can make certain laws.")
```

## üë• Population Powerhouses: Most and Least Populous States

India is home to over 1.4 billion people! Let's see how they're distributed across different states.

```python
# Sort states by population
states_only = india_df[india_df['Type'] == 'State'].copy()
states_sorted = states_only.sort_values('Population_Millions', ascending=False)

# Top 10 most populous states
top_10 = states_sorted.head(10)
bottom_5 = states_sorted.tail(5)

print("üèÜ TOP 10 MOST POPULOUS STATES")
print("=" * 35)
for i, (_, row) in enumerate(top_10.iterrows(), 1):
    print(f"{i:2d}. {row['State/UT']:<20} - {row['Population_Millions']:>6.1f} million")

print("\nüå± 5 LEAST POPULOUS STATES")
print("=" * 30)
for i, (_, row) in enumerate(bottom_5.iterrows(), 1):
    print(f"{i}. {row['State/UT']:<20} - {row['Population_Millions']:>6.1f} million")

# Create visualization
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 12))

# Top 10 states bar chart
bars1 = ax1.barh(range(len(top_10)), top_10['Population_Millions'], 
                 color=plt.cm.Reds(np.linspace(0.4, 0.8, len(top_10))))
ax1.set_yticks(range(len(top_10)))
ax1.set_yticklabels(top_10['State/UT'])
ax1.set_xlabel('Population (Millions)')
ax1.set_title('Top 10 Most Populous States', fontsize=14, fontweight='bold')
ax1.invert_yaxis()

# Add value labels
for i, bar in enumerate(bars1):
    width = bar.get_width()
    ax1.text(width + 1, bar.get_y() + bar.get_height()/2, 
             f'{width:.1f}M', ha='left', va='center', fontweight='bold')

# Bottom 5 states bar chart
bars2 = ax2.barh(range(len(bottom_5)), bottom_5['Population_Millions'], 
                 color=plt.cm.Blues(np.linspace(0.4, 0.8, len(bottom_5))))
ax2.set_yticks(range(len(bottom_5)))
ax2.set_yticklabels(bottom_5['State/UT'])
ax2.set_xlabel('Population (Millions)')
ax2.set_title('5 Least Populous States', fontsize=14, fontweight='bold')
ax2.invert_yaxis()

# Add value labels
for i, bar in enumerate(bars2):
    width = bar.get_width()
    ax2.text(width + 0.05, bar.get_y() + bar.get_height()/2, 
             f'{width:.1f}M', ha='left', va='center', fontweight='bold')

plt.tight_layout()
plt.show()

# Fun facts
largest_state = top_10.iloc[0]
smallest_state = bottom_5.iloc[-1]

print(f"\nü§Ø Amazing Facts:")
print(f"‚Ä¢ {largest_state['State/UT']} has {largest_state['Population_Millions']:.1f} million people!")
print(f"‚Ä¢ That's {largest_state['Population_Millions']/smallest_state['Population_Millions']:.0f} times more than {smallest_state['State/UT']}!")
print(f"‚Ä¢ If {largest_state['State/UT']} were a country, it would be the 5th most populous in the world!")
```
