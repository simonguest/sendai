```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
x = np.linspace(0, 10, 100)
y = np.sin(x) + np.random.normal(0, 0.1, 100)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, 'b-', label='Random Data')
plt.title('Simple Matplotlib Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```
